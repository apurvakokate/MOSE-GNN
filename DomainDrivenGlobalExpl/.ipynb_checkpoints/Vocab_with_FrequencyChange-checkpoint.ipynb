```python
import csv
from collections import Counter, defaultdict
from Utils_vocab import *
import math

def process_and_save_motif_data(dataset_names, least_counts, dataset_column_dict):
    '''
    Note not using k-folds
    '''
    
    for algorithm in ["RBRICS", "MGSSL"]: #"BRICS"
        output_csv=f"motif_analysis_{algorithm}.csv"
        results = []
        for dataset_name in dataset_names:
            for least_count in least_counts:

                print(dataset_name, least_count, algorithm)
                # Load datasets
                training_data = MolDataset(
                    root=".", split='training',
                    csv_file=f"datasets/_{dataset_name}.csv", 
                    label_col=dataset_column_dict[dataset_name]
                )
                validation_data = MolDataset(
                    root=".", split='valid',
                    csv_file=f"datasets/_{dataset_name}.csv", 
                    label_col=dataset_column_dict[dataset_name]
                )
                test_data = MolDataset(
                    root=".", split='test',
                    csv_file=f"datasets/_{dataset_name}.csv", 
                    label_col=dataset_column_dict[dataset_name]
                )

                data_lookup, test_data_lookup, lookup_object = get_lookup_tables(training_data, validation_data, test_data, algorithm = algorithm)

                value_counts = Counter(
                    motif for graph_data in data_lookup.values() for motif in graph_data.values()
                )
                value_counts_test = Counter(
                    motif for graph_data in test_data_lookup.values() for motif in graph_data.values()
                )

                for motif, length in lookup_object.motifs_length.items():
                    value_counts[motif] /= length
                for motif, length in lookup_object.test_motifs_length.items():
                    value_counts_test[motif] /= length


                # Identify motifs to remove
                motifs_to_remove = {motif for motif, count in value_counts.items() if count <= least_count}
                lookup_object.remove_motifs(list(motifs_to_remove))

                # Reindex motifs and compute node coverage
                motif_list = list(lookup_object.get_all_unique_motif())

                train_val_coverage, graph_to_motifs = reindex_data(data_lookup, motif_list, value_counts, least_count)
                test_coverage, test_graph_to_motifs = reindex_data(test_data_lookup, motif_list, value_counts_test, least_count,is_test=True)

                # Compute mean length
                total_weighted_length_train_val = sum(
                    value_counts[key] * length
                    for key, length in lookup_object.motifs_length.items()
                    if key in value_counts
                )

                total_count_train_val = sum(value_counts.values())

                mean_length_train_val = total_weighted_length_train_val / total_count_train_val if total_count_train_val != 0 else 0
                # Compute variance
                variance = sum(
                    value_counts[key] * (length - mean_length_train_val) ** 2
                    for key, length in lookup_object.motifs_length.items()
                    if key in value_counts
                ) / total_count_train_val if total_count_train_val != 0 else 0

                # Compute standard deviation
                std_dev_train = math.sqrt(variance)

                total_weighted_length_test = sum(
                    value_counts_test[key] * length
                    for key, length in lookup_object.motifs_length.items()
                    if key in value_counts_test
                )

                total_count_test = sum(value_counts_test.values())

                mean_length_test = total_weighted_length_test / total_count_test if total_count_test != 0 else 0

                # Compute variance
                variance = sum(
                    value_counts[key] * (length - mean_length_test) ** 2
                    for key, length in lookup_object.motifs_length.items()
                    if key in value_counts_test
                ) / total_count_test if total_count_test != 0 else 0

                # Compute standard deviation
                std_dev_test = math.sqrt(variance)

                # Save results
                results.append({
                    "dataset_name": dataset_name,
                    "least_count": least_count,
                    "total_graphs_train_val": len(data_lookup),
                    "total_graphs_test": len(test_data_lookup),
                    "vocabulary_size": len(motif_list),
                    "graphs_with_motifs_train_val": len(graph_to_motifs),
                    "graphs_with_motifs_test": len(test_graph_to_motifs),
                    "node_coverage_train_val": train_val_coverage,
                    "node_coverage_test": test_coverage,
                    "motif_length_train": f"{mean_length_train_val} +- {std_dev_train}",
                    "motif_length_test": f"{mean_length_test} +- {std_dev_test}",


                })

        # Write results to a CSV file
        with open(output_csv, mode="w", newline="") as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=results[0].keys())
            writer.writeheader()
            writer.writerows(results)

        print(f"Results saved to {output_csv}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data into a pandas DataFrame

def plot_csv(path='motif_analysis_BRICS.csv'):
    # Replace 'your_file.csv' with the path to your CSV file
    df = pd.read_csv(path)

    # Normalize graphs_with_motifs counts
    df['train_val_fraction'] = df['graphs_with_motifs_train_val'] / df['total_graphs_train_val']
    df['test_fraction'] = df['graphs_with_motifs_test'] / df['total_graphs_test']

    # Initialize a plot
    fig, axs = plt.subplots(1, 2, figsize=(15, 6), sharex=True)

    # First subplot: Node Coverage
    for dataset in df['dataset_name'].unique():
        subset = df[df['dataset_name'] == dataset]
        axs[0].plot(
            subset['least_count'], 
            subset['node_coverage_train_val'], 
            label=f"{dataset} - Train/Val", 
            marker='o'
        )
        # Annotate points
        for _, row in subset.iterrows():
            axs[0].annotate(f"{row['node_coverage_train_val']:.1f}", (row['least_count'], row['node_coverage_train_val']), fontsize=8)
        axs[0].plot(
            subset['least_count'], 
            subset['node_coverage_test'], 
            label=f"{dataset} - Test", 
            marker='x'
        )
        # Annotate points
        for _, row in subset.iterrows():
            axs[0].annotate(f"{row['node_coverage_test']:.1f}", (row['least_count'], row['node_coverage_test']), fontsize=8)

    axs[0].set_title("Node Coverage")
    axs[0].set_xlabel("Least Count")
    axs[0].set_ylabel("Node Coverage")
    axs[0].legend()

    # Second subplot: Fraction of Graphs with Motifs
    for dataset in df['dataset_name'].unique():
        subset = df[df['dataset_name'] == dataset]
        axs[1].plot(
            subset['least_count'], 
            subset['train_val_fraction'], 
            label=f"{dataset} - Train/Val", 
            marker='o'
        )
        # Annotate points
        for _, row in subset.iterrows():
            axs[1].annotate(f"{row['train_val_fraction']:.1f}", (row['least_count'], row['train_val_fraction']), fontsize=8)
        axs[1].plot(
            subset['least_count'], 
            subset['test_fraction'], 
            label=f"{dataset} - Test", 
            marker='x'
        )
        # Annotate points
        for _, row in subset.iterrows():
            axs[1].annotate(f"{row['test_fraction']:.1f}", (row['least_count'], row['test_fraction']), fontsize=8)

    axs[1].set_title("Fraction of Graphs with Motifs")
    axs[1].set_xlabel("Least Count")
    axs[1].set_ylabel("Fraction of Graphs")
    axs[1].legend()

    # Show plot
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator


def plot_csv(path='motif_analysis_BRICS.csv'):
    # Load your data into a pandas DataFrame
    df = pd.read_csv(path)

    # Normalize graphs_with_motifs counts
    df['train_val_fraction'] = df['graphs_with_motifs_train_val'] / df['total_graphs_train_val']
    df['test_fraction'] = df['graphs_with_motifs_test'] / df['total_graphs_test']

    # Initialize a plot
    fig, axs = plt.subplots(1, 3, figsize=(15, 6), sharex=True)

    # Define colors for datasets
    colors = plt.cm.tab10(range(len(df['dataset_name'].unique())))

    # First subplot: Node Coverage
    for i, dataset in enumerate(df['dataset_name'].unique()):
        subset = df[df['dataset_name'] == dataset]
        train_val_line, = axs[0].plot(
            subset['least_count'], 
            subset['node_coverage_train_val'], 
            label=f"{dataset} - Train/Val", 
            marker='o', 
            color=colors[i]
        )
        test_line, = axs[0].plot(
            subset['least_count'], 
            subset['node_coverage_test'], 
            label=f"{dataset} - Test", 
            linestyle='--', 
            marker='x', 
            color=colors[i]
        )
        # Annotate points with coverage values
        for row_idx, row in subset.iterrows():
            if row_idx % 2 == 0:  # Annotate every 4th point
                axs[0].annotate(
                    f"{row['node_coverage_train_val']:.2f}", 
                    (row['least_count'], row['node_coverage_train_val']), 
                    fontsize=8
                )
                axs[0].annotate(
                    f"{row['node_coverage_test']:.2f}", 
                    (row['least_count'], row['node_coverage_test']), 
                    fontsize=8
                )

    axs[0].set_title("Node Coverage")
    axs[0].set_xlabel("Least Count")
    axs[0].set_ylabel("Node Coverage")

    # Second subplot: Fraction of Graphs with Motifs
    for i, dataset in enumerate(df['dataset_name'].unique()):
        subset = df[df['dataset_name'] == dataset]
        axs[1].plot(
            subset['least_count'], 
            subset['train_val_fraction'], 
            label=f"{dataset} - Train/Val", 
            marker='o', 
            color=colors[i]
        )
        axs[1].plot(
            subset['least_count'], 
            subset['test_fraction'], 
            label=f"{dataset} - Test", 
            linestyle='--', 
            marker='x', 
            color=colors[i]
        )
        # Annotate points with fraction values
        for row_idx, row in subset.iterrows():
            if row_idx % 2 == 0:  # Annotate every 4th point
                axs[1].annotate(
                    f"{row['train_val_fraction']:.2f}", 
                    (row['least_count'], row['train_val_fraction']), 
                    fontsize=8
                )
                axs[1].annotate(
                    f"{row['test_fraction']:.2f}", 
                    (row['least_count'], row['test_fraction']), 
                    fontsize=8
                )

    axs[1].set_title("Fraction of Graphs with Motifs")
    axs[1].set_xlabel("Least Count")
    axs[1].set_ylabel("Fraction of Graphs")
    print(df['motif_length_train'])
    
    df['motif_length_train_mean'] = df['motif_length_train'].str.split(r' \+- ').str[0].astype(float)
    df['motif_length_train_std'] = df['motif_length_train'].str.split(r' \+- ').str[1].astype(float)
    df['motif_length_test_mean'] = df['motif_length_test'].str.split(r' \+- ').str[0].astype(float)
    df['motif_length_test_std'] = df['motif_length_test'].str.split(r' \+- ').str[1].astype(float)
    
    # Second subplot: Fraction of Graphs with Motifs
    for i, dataset in enumerate(df['dataset_name'].unique()):
        subset = df[df['dataset_name'] == dataset]

        axs[2].errorbar(
            subset['least_count'], 
            subset['motif_length_train_mean'], 
            yerr=subset['motif_length_train_std'], 
            label=f"{dataset} - Train/Val", 
            marker='o', 
            alpha = 0.25,
            color=colors[i]
        )
        axs[2].errorbar(
            subset['least_count'], 
            subset['motif_length_test_mean'], 
            yerr=subset['motif_length_test_std'], 
            label=f"{dataset} - Test", 
            linestyle='--', 
            marker='x', 
            alpha = 0.25,
            color=colors[i]
        )

    # Adjust y-axis ticks to show approximately 10 values
    axs[2].yaxis.set_major_locator(MaxNLocator(nbins=10))

    axs[2].set_title("Fraction of Graphs with Motifs")
    axs[2].set_xlabel("Least Count")
    axs[2].set_ylabel("Fraction of Graphs")
    axs[2].legend()

    # Combine legend for all datasets
    handles, labels = axs[0].get_legend_handles_labels()
    fig.legend(
        handles=handles,
        labels=labels,
        loc='center right', 
        bbox_to_anchor=(1.15, 0.5),
        title="Dataset Coverage"
    )

    # Adjust layout
    plt.tight_layout(rect=[0, 0, 0.9, 1])
    plt.show()
```
